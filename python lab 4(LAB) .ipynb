### EVAL() AND QUERY()

```python
import numpy as np
import pandas as pd
```

```python
nrows,ncols=100000,100
rng=np.random.RandomState(42)
df1,df2,df3,df4=(pd.DataFrame(rng.rand(nrows,ncols))
for i in range(4)) # the for loop is for reduce the code for creating the rng
```

```python
print(df2)
```

```python
%timeit df1+df2+df3+df4
```

```python
%timeit pd.eval('df1+df2+df3+df4') # it reduces the time of execution it must be given in sngle quotes
```

```python
np.allclose((df1+df2+df3+df4),pd.eval('df1+df2+df3+df4')) # if both are ture it returns true
```

```python
df1,df2,df3,df4=(pd.DataFrame(rng.randint(0,100,(10,2)))
for i in range(4))
```

```python
df1
```

```python
df2
```

```python
res1= df1+df2/df3*df4
res2=pd.eval('df1+df2/df3*df4')
```

```python
np.allclose(res1,res2)
```

```python
%timeit res1
```

```python
%timeit res2
```

```python
res1= df1+df2/df3+df4
res2=pd.eval('df1+df2/df3*df4')
```

```python
np.allclose(res1,res2)
```

```python
# bitewise operator
```

```python
res3=(df1<0.5)&(df2<0.5)|(df3<df4)
res4=pd.eval('(df1<0.5) and (df2<0.5) or (df3<df4)')
np.allclose(res3,res4)
```

```python
%timeit res3
```

```python
%timeit res4
```

```python
# column operration
```

```python
df=pd.DataFrame(rng.rand(100,3),columns=['A','B','C'])
df.head()
```

```python
re1=(df['A']+df['B']/df['C']-1)
re2=pd.eval("(df.A+df.B)/(df.C-1)")
np.allclose(re1,re2)
```

```python
# new col
```

```python
df.eval('D=(A+B)/C',inplace=False)
```

```python
col_mean=df.mean(1)# row mean
res1=df['A']+col_mean
res2=df.eval('A+@col_mean')
np.allclose(res1,res2)
```

```python
ax=plt.axes(projection='3d')
xline=np.linspace(0,15,1000)
yline=np.sin(xline)
zline=np.cos(xline)
ax.plot3D(xline,yline,zline,'blue')
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
x,y=np.meshgrid(x,y)
z=f(x,y)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(x,y,z,30,cmap='plasma')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
```

```python
ax.view_init(-30,-15)
fig
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_wireframe(x,y,z,color='black')
ax.set_title('Wireframe')
```

```python
ax.view_init(-30,-15)
fig
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap='gray',edgecolor='white')
ax.set_title('surface')
```
