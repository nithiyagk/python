```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line graph of sin')
```

```python
x = np.linspace(0,10,30)
y = np.cos(x)
plt.plot(x,y,marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line graph of cos')
```

```python
x = np.linspace(0,10,30)
y = np.tan(x)
plt.plot(x,y,marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line graph of tan')
```
