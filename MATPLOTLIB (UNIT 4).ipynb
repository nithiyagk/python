```python
import matplotlib.pyplot as plt # refered in net

# Data for the pie chart
labels = ['APPLE', 'WATERMELON', 'LEMON', 'MANGO']
sizes = [30, 25, 20, 25]
colors = ['skyblue', 'lightgreen', 'orange', 'pink']

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Pie Chart Example')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
```

```python
lables = ['APPLE','ORANGE','WATERMELON','MANGO']
sizes = [20,26,52,12]  # it must have only 100 % of value in the sizes or in lables
colors = ['blue', 'green', 'red', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=30,shadow = True)
plt.axis('equal')
plt.title('SALES OF FRUITS IN A YEAR')
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.2,0.5,0.5])
ax3 = fig.add_axes([0.6,0.6,0.5,0.5])

ax1.plot(x,y)
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
ax1.set_title('plotting1')

# axes 2
ax2.plot(x,y**2,'r')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')
ax2.set_title('plotting2')

# axes3
ax3.plot(x,y,'g')
ax3.set_xlabel('x-axis')
ax3.set_ylabel('y-axis')
ax3.set_title('plotting3')
```

```python
fig,ax=plt.subplots(1,3,figsize=(10,10)) # can also change the axis by 3,1
ax[0].plot(x,y,'m*')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('plot1')

ax[1].plot(x,y**2,'c-.')
ax[1].set_xlabel('x1')
ax[1].set_ylabel('y1')
ax[1].set_title('plot2')

ax[2].plot(x,y,'r--')
ax[2].set_xlabel('x2')
ax[2].set_ylabel('y2')
ax[2].set_title('plot1')

plt.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=1)
```

```python
rng = np.random.RandomState(0)
x= rng.randn(100)
y= rng.randn(100)
colors = rng.randn(100)
size = 1000*rng.rand(100)
plt.scatter(x,y,c= colors,s=size,alpha=1.0,cmap='viridis') #alpha is used to increase or decrease the drkness of the colors and its amximum value is 1
plt.colorbar(); # i is used to show the color bar 
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T # T --> transpose matrix
plt.scatter(features[0],features[1],alpha = 0.8,s =100*features[3],c= iris.target,cmap = 'viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar();
```
